It show that outliers are present in dataset or not we have outliers present in weendspeed featutre we neet to deal with that

```python
# Calculate Quartiles and IQR
Q1 = df.windspeed.quantile(0.25)
Q3 = df.windspeed.quantile(0.75)
IQR = Q3 - Q1
# Outlier Filtering
filtered_data = df.windspeed[~((df.windspeed < (Q1 - 1.5 * IQR)) | (df.windspeed > (Q3 + 1.5 * IQR)))]
```

This code calculates the upper bound for outliers in the 'windspeed' column using the IQR method and prints the upper bound value along with the count of values exceeding this upper bound.

```python
# Calculate upper bound for outliers
upper = Q3+1.5*IQR
# Create a boolean array identifying values exceeding the upper bound
upper_array = np.array(df['windspeed'] >= upper)
# Print upper bound and count of values exceeding it
print("Upper Bound:", upper)
print(upper_array.sum())
```

This code calculates the lower bound for outliers in the 'windspeed' column using the IQR method and prints the lower bound value along with the count of values below this lower bound

```python
# Calculate lower bound for outliers
lower = Q1-1.5*IQR
# Create a boolean array identifying values below the lower bound
lower_array = np.array(df['windspeed'] <= lower)
# Print lower bound and count of values below it
print("Lower Bound:", lower)
print(lower_array.sum())
```

```python
''' Detection '''
# IQR
# Calculate the upper and lower limits
Q1 = df['windspeed'].quantile(0.25)
Q3 = df['windspeed'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
# Create arrays of Boolean values indicating the outlier rows
upper_array = np.where(df['windspeed'] >= upper)[0]
lower_array = np.where(df['windspeed'] <= lower)[0]
```

```python
# Removing the outliers
df.drop(index=upper_array, inplace=True)
df.drop(index=lower_array, inplace=True)
```

## Spliting data into training and testing data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

## Model Selection

This code snippet uses the scikit-learn library to train a RandomForestClassifier on training data (x_train and y_train) and then make predictions on test data (x_test)

```python
# Import the RandomForestClassifier from scikit-learn
from sklearn.ensemble import RandomForestClassifier
# Create an variable of the RandomForestClassifier
rfc = RandomForestClassifier()
# Train the RandomForestClassifier on the training data
rfc.fit(x_train,y_train)
# Make predictions on the test data
y_pred = rfc.predict(x_test)
```

```python
y_pred
```

```python
# Import the r2_score function from scikit-learn metrics
from sklearn.metrics import r2_score
# Calculate the R-squared accuracy score
accuracy = r2_score(y_test, y_pred)
```

The r2_score function from scikit-learn to calculate the R-squared (coefficient of determination) metric to assess the accuracy of predictions
